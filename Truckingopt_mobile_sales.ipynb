```python
cd = data.copy()
```

```python
#cleaning data
cd.fillna(round(cd.mean(),2), inplace=True)
```

# **Exploratory Data Analysis**

```python
# Plot a heatMap
plt.figure(figsize=(20,10))
sns.heatmap(cd.corr(),annot=True)
```

```python
# Check distribution of Target Variable
sns.distplot(cd.Sold)
plt.title("Average Number of mobile phone sold",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
cd.groupby(['City'])['Sold'].count().sort_values(ascending=False).head(10).plot.bar()
plt.ylabel('No of mobile sold')
plt.title("Top cities where sale is highest",fontsize=20)
plt.show()
```

```python
cd.groupby(['State'])['Sold'].count().sort_values(ascending=False).head(10).plot.bar()
plt.ylabel('No of mobile sold')
plt.title("Top states where sale is highest",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
cd.groupby(['location_name'])['Sold'].count().sort_values(ascending=False).head(10).plot.bar()
plt.ylabel('No of mobile sold')
plt.title("Top location where sale is highest",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
cd.groupby(['City'])['raw_visitor_counts'].count().sort_values(ascending=False).head(10).plot.bar()

plt.title("Top cities having highest visitor",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
cd.groupby(['distance_from_home'])['Sold'].count().sort_values(ascending=False).head(10).plot.bar()
plt.ylabel('No of mobile sold')
plt.title("distance from home comparision",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
cd.groupby(['device_type'])['Sold'].count().sort_values(ascending=False).head(10).plot.bar()
plt.ylabel('No of mobile sold')
plt.title("Top devices whose sale is highest",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,5))
cd.groupby(['open_hours'])['Sold'].count().sort_values(ascending=False).head(10).plot.bar()
plt.ylabel('number of mobile sold')
plt.title("Most profitable hours",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,5))
cd.groupby(['popularity_by_hour'])['Sold'].count().sort_values(ascending=False).head(10).plot.bar()
plt.ylabel('number of mobile sold')

plt.show()
```

```python
plt.figure(figsize=(30,40))
sns.pairplot(cd)
plt.show()
```
